# Traditional feature-based Methods: Node

### Node Feature란,

각 노드가 가진 속성 값 또는 정보 벡터
예)

- 사람 노드: 나이, 성별, 직업, 관심사 등
- 논문 노드: 제목, 단어 빈도, 분야 등
- 물질 노드: 원자 번호, 전하, 질량 등

- Graphlets counting은 그래프에서 작은 부분 서브그래프(subgraph) 들의 패턴 빈도를 세는 과정으로, 그래프의 구조적 특성을 분석할 때 매우 중요한 도구이다.
