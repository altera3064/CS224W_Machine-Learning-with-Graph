### - 그래프에서 가깝고 더 중요할 수록, 더 많이 방문한다는 것을 알 수 있다.

- 가깝다 = 출발 노드로부터 짧은 거리에 있는 노드

- 더 중요하다 = 연결(edge)이 많거나, 연결된 이웃이 많은 노드

### - 초록색은 공통된 이웃이 있고, 다른 하나로 이어지는 경로가 많다는 것을 의미한다.

- 초록색 노드는 공통된 이웃을 많이 가진 노드.

- 예를 들어, 유저 A와 B가 모두 핀 X를 저장했다면,
→ 핀 X는 A와 B 모두에게 연결되어 있는 공통 이웃이에요.

- 이런 노드는 두 유저 사이에서 경로가 많이 존재하게 됩니다

### - 이러한 경로는 짧기 때문에, 무작위 워커가 다시 시작하기로 결정하지 않는 식으로 진행된다.

- 만약 현재 경로가 짧고 유망한 경로라면,
- teleport 없이 계속 걷게 되는 확률이 높아진다.
