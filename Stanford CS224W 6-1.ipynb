- 그래프의 노드들을 d차원으로 매핑해서,
- 그래프에서 비슷한 노드가 서로 가깝게 임베딩 되도록 하는 것이었다.

- 결론적으로, 그래프를 입력으로 받아서 각 노드의 위치와 임베딩을 알려주는 함수 f를 배우는 것이었다.

image.png

- 우리는 인코더-디코더 프레임워크를 생각했다.
- 노드의 유사성이 임베딩 공간에서 유사하거나 일치하기를 원했다.
- 유사성을 통해서 각 노드끼리의 거리를 측정할 수 있었다.
- 측정할 수 있는 거리에는 내적(dot product)과 코사인 거리가 있었다.

image.png

- Encoder : 노드들을 좌표공간(vector space)으로 바꾸는 것
- Decoder : 이 좌표들을 이용해 두 노드가 유사한지를 알아보는 것이 디코더

- Similarity function : 실제 네트워크의 관계를 벡터 공간에서의 관계를 특정하는 함수
- 어떤 기준으로 유사하다고 할지 정하는 것.
- z x z 는 Decoder이다.

### - O(V) 파라미터가 필요




- 각각의 노드가 서로 공유하지 않는다.
- 각 노드는 고유한 임베딩 벡터를 가진다.

### - 변환적(Trasductive)


- 훈련 단계에서 실제로 본 사례에 대해서만 예측을 할 수 있다는 것을 의미한다.

### - 노드 기능을 통합하지 않다.

- 실제 네트워크에는 노드에 연결된 기능, 속성이 있다.
- 이런 접근 방식은 자연스럽게 이를 활용하지 못한다.
